```python
import cv2
import json
import os
from glob import glob

import numpy as np
from PIL import Image

data_root = "c:/Users/Noh/github/Accident_Prediction_Prevent/Models/Object_Detection/sample_data/"

json_dir = data_root + "test_data/라벨링데이터_231108_add/차대보행자_이미지_육교및지하도부근/bb_1_170402_pedestrian_112_329/" # 모든 JSON 파일이 저장된 폴더
source_dir = data_root + "test_data/원천데이터_231108_add/차대보행자_이미지_육교및지하도부근/bb_1_170402_pedestrian_112_329" # 원본 이미지 파일이 저장된 폴더

output_json_file = data_root + "coco_annotations - bb_1_170402_pedestrian_112_329.json"
```

```python
def visualize_predictions(image_path, annotations, output_path=None):
    # Read image
    # image = cv2.imread(str(image_path))
    image = Image.open(image_path).convert("RGB")  # PIL로 로드
    image = np.array(image)  # OpenCV와 호환되도록 변환
    
    # Colors for different categories
    colors = {
    0: (128, 0, 128),   # 보라색 - two-wheeled-vehicle
    1: (255, 0, 0),     # 빨간색 - traffic-light-red
    2: (0, 255, 0),     # 초록색 - pedestrian
    3: (255, 255, 0),   # 노란색 - crosswalk
    4: (0, 255, 255),   # 하늘색 - bike
    5: (0, 128, 0),     # 짙은 초록 - traffic-light-green
    6: (0, 0, 255),     # 파란색 - vehicle
    7: (255, 165, 0),   # 주황색 - traffic-light-etc
    8: (255, 255, 255)  # 흰색 - traffic-sign
}

    
    # Draw bounding boxes
    for ann in annotations:
        
        bbox = ann["bbox"]
        cat_id = ann["category_id"]
        
        # Draw rectangle
        x, y, w, h = [int(coord) for coord in bbox]
        cv2.rectangle(image, (x, y), (x+w, y+h), colors[cat_id], 2)
        
        # Add category and score text
        text = f"cat:{cat_id} {ann['score']:.2f}"
        cv2.putText(image, text, (x, y-5), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[cat_id], 2)

    # Create output directory if it doesn't exist
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    
    # Save the image
    cv2.imwrite(str(output_path), image)
    print(f"Saved annotated image to: {output_path}")
```

```python
# Load predictions
json_files = glob(os.path.join(json_dir, "*.json"))  # 모든 JSON 파일 검색

pred_path = "c:/Users/Noh/github/Accident_Prediction_Prevent/Models/Object_Detection/datasplit/"

with open(pred_path + 'pred.json', 'r') as f:
    predictions = json.load(f)

with open(output_json_file, 'r') as f:
    annotations = json.load(f)

source_img = {}
for ann in annotations["images"]:
    image_id = ann["id"]
    source_img_path = ann["file_name"]
    if image_id not in source_img:
        source_img[image_id] = []

    source_img[image_id].append(source_img_path)

# Group annotations by image_id
pred_by_image = {}
for ann in predictions:
    image_id = ann["image_id"]
    if image_id not in pred_by_image:
        pred_by_image[image_id] = []
    pred_by_image[image_id].append(ann)

# Directory containing images
image_dir = source_dir

# Create output directory
output_dir = os.path.join(pred_path, "visualized_predictions" + "bb_1_170402_pedestrian_112_329")
os.makedirs(output_dir, exist_ok=True)

# Visualize each image
for image_id in pred_by_image:
    image_path = source_img[image_id][0]

    if os.path.exists(image_path):
        output_path = os.path.join(output_dir, f"{image_id}_pred.png")
        visualize_predictions(image_path, pred_by_image[image_id], output_path)
```
